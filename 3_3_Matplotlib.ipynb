Matplotlib:


*   Useful for making Plots



```python
# importing matplotlib
import matplotlib.pyplot as plt
```

```python
# import numpy to get data for our plots
import numpy as np
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)
z = np.cos(x)
```

```python
print(x)
```

```python
print(y)
```

```python
print(z)
```

Ploting the data

```python
# sine wave

plt.plot(x,y)
plt.show()

# cos wave
plt.plot(x,z)
plt.show()
plt.plot(y,z)
plt.show()
plt.plot(x,y,z)
plt.show()
```

Adding title x-axis and y-axis labels

```python
plt.plot(x,y)
plt.xlabel('angle')
plt.ylabel('sine value')
plt.title('Sine Wave')
plt.show()
```

Bar Plot

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
languages = ['English', 'French', 'Spanish', 'Latin', 'German']
peoples = [100, 50, 150, 40, 70]
ax.bar(languages,peoples)
plt.xlabel('LANGUAGES')
plt.ylabel('NUMBER OF PEOPLES')
plt.title('Bar Graph')
plt.show()
```

Pie Chart

```python
fig1 = plt.figure()
ax = fig1.add_axes([0,0,1,1])
languages = ['English', 'French', 'Spanish', 'Latin', 'German']
peoples = [100, 50, 150, 40, 70]
plt.pie(peoples,labels=languages, autopct='%1.1f%%')
plt.title('Pie Chart')
plt.show()
```

```python
x = np.linspace(0,10,30)
y = np.sin(x)
z = np.cos(x)

fig2 = plt.figure()
ax = fig2.add_axes([0,0,1,1])
ax.scatter(x,y,color='g')
ax.scatter(x,z,color='b')
plt.title('Scatter Graph')
plt.show()
```

3D Scatter Plot

```python
fig3 = plt.figure()
ax = plt.axes(projection='3d')
z = 20 * np.random.random(100)
x = np.sin(z)
y = np.cos(z)
ax.scatter(x,y,z,c=z,cmap='viridis')
plt.title('3D graph')
plt.show()
```
