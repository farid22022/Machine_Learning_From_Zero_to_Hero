# Operator

1. Arithmatic

```python
num_1 = 20
num_2 = 30

#addition
sum = num_1 + num_2
print('sum of {0} and {1} is {2}'.format(num_1, num_2, sum))

#substraction
diff = num_2 - num_1
print('difference of {0} and {1} is {2}'.format(num_1, num_2, diff))

#multiplication
product = num_1 * num_2
print('product of {0} and {1} is {2}'.format(num_1, num_2, product))


#divison
div = num_2 / num_1
print('divison of {0} and {1} is {2}'.format(num_1, num_2, div))

#exponenet
exp = num_1 ** num_2
print('{0} to the power {1} is {2}'.format(num_1, num_2, exp))

#modulus
mod = num_2 % num_1
print('modulus or reminder of {0} and {1} is {2}'.format(num_1, num_2, mod))
```

# Function

```python
number = int(input(" Enter a number to know it's factorial: "))

factorial =1

if number == 0:
  print(" The factoria of 0 is : 1")
else:
  for i in range(1,number+1):
    factorial = factorial * i
  print(factorial)
```

```python
def factorial_value(num):
  factorial =1

  if num == 0:
    return factorial
  else:
    for i in range(1,num+1):
      factorial = factorial * i
    return factorial
```
